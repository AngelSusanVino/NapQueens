   Region: Central seems to be the dominant region.

   Category: Office Supplies appears to be the most frequent category.

   Sub-Category: Fasteners seems to be the most common subcategory within Office Supplies.

   Order Priority: Medium priority orders appear to be the most frequent.

#### Correlation

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

Sales and Quantity: A positive correlation (0.31), indicating that higher sales are often associated with larger quantities.

Sales and Profit: A moderate positive correlation (0.48) suggests that higher sales tend to lead to higher profits.

Shipping Cost and Sales: A strong positive correlation (0.77) indicates that higher shipping costs are often associated with higher sales.

Discount and Profit: A moderate negative correlation (-0.32) suggests that higher discounts tend to lead to lower profits.

Sales vs Profit - A positive correlation exists, suggesting higher sales lead to higher profits.

#### Time Series Analysis

##### Sales Over Time

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date')['Sales'].resample('M').sum().plot(figsize=(12, 6))
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

We have data from Jan 2011 - End of 2014.

There are significant ups and downs in sales over the period.

As per this we can say there might me a seaseonal trend that is existing.

Even if there are  fluctuations, there seems to be an overall upward trend in sales towards the end of the period.


##### Profit Over Time

```python
df.set_index('Order Date')['Profit'].resample('M').sum().plot(figsize=(12, 6))
plt.title('Monthly Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.show()
```

There exist significant ups and downs in profit throughout the period.

Unlike the sales plot, there isn't a clear upward trend in profit. 

The profit seems to fluctuate around a certain level with periods of higher and lower profits.

#### Sales and Profit By Cateogory

```python
# Group by Category and calculate total Sales, Profit, and Order Count
category_sales_profit = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum', 'Order ID': 'count'}).reset_index()
category_sales_profit.rename(columns={'Order ID': 'Order Count'}, inplace=True)

fig = px.scatter(category_sales_profit, x='Sales', y='Profit', size='Order Count', color='Category',
                 hover_name='Category', title='Categories by Sales and Profit',
                 labels={'Sales': 'Total Sales', 'Profit': 'Total Profit', 'Order Count': 'Number of Orders'},
                 size_max=60, color_discrete_sequence=px.colors.qualitative.Vivid)

fig.show()
```

Here we can see Technology is the one which shows as highest sales and highest profit.

#### Sales and Profit by Sub Category

```python
# Group by Sub-Category and calculate total Sales, Profit, and Order Count
subcategory_sales_profit = df.groupby('Sub-Category').agg({'Sales': 'sum', 'Profit': 'sum', 'Order ID': 'count'}).reset_index()
subcategory_sales_profit.rename(columns={'Order ID': 'Order Count'}, inplace=True)

fig = px.scatter(subcategory_sales_profit, x='Sales', y='Profit', size='Order Count', color='Sub-Category',
                 hover_name='Sub-Category', title='Sub-Categories by Sales and Profit',
                 labels={'Sales': 'Total Sales', 'Profit': 'Total Profit', 'Order Count': 'Number of Orders'},
                 size_max=60, color_discrete_sequence=px.colors.qualitative.Vivid)

fig.show()
```

Here we can observe phones have comparitatively higher sales and and profit but copiers do show higher profit levels than in sales when compared with phones.
